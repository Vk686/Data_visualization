```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,20)
```

```python
y = x + x  
z = x * x
```

```python
y
```

```python
plt.plot(x,z)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Table')
```

```python
plt.subplot(3,2,1)
plt.plot(y,z)
plt.subplot(3,2,3)
plt.plot(y,x)
plt.subplot(3,2,2)
plt.plot(x,z)
plt.subplot(3,2,4)
plt.plot(x,y)
plt.subplot(3,2,6)
plt.plot(x,x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.tight_layout()
```

```python
plt.bar(x,x)
```

```python
plt.subplot(3,2,1)
plt.bar(x,x)
plt.subplot(3,2,4)
plt.fill(x,z)
plt.subplot(3,2,5)
plt.hist(x,y)
plt.subplot(3,2,3)
plt.plot(y,z)
plt.tight_layout()
```

```python
plt.polar(x,z)
```
