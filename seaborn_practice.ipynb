```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flight2 = flights.corr()
```

```python
flight2
```

```python
sns.heatmap(flight2)
```

```python
sns.heatmap(flight2, annot = True, linewidth = 3, linecolor = 'blue', cmap = 'coolwarm')
```

```python
flight2 = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
flight2
```

```python
sns.heatmap(flight2, lw = 3, cmap = 'cool')
```

```python
sns.clustermap(flight2, lw = 3, cmap = 'cool')
```
